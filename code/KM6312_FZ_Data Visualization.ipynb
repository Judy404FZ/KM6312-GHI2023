```python
#Wordcloud of Countries

unique_countries = df['Country Name'].drop_duplicates()
unique_countries = unique_countries.str.replace(' ', '_')
text = ' '.join(unique_countries)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  
plt.show()
```

```python
# Bar plot of regions
plt.bar(x = df['region'].value_counts().index, 
        height = df['region'].value_counts().values,
         color = 'g', edgecolor = 'k',align='center')

plt.xticks(rotation=45,ha='right')
plt.xlabel('Regions'); plt.ylabel('Count'); plt.title('Distribution of Regions');
```

```python
df_Singapore=df[df['Country Name']=='Singapore']
df_Singapore
```

```python
# Bar plot of happiness score of Singpaore

#计算每年的平均幸福指数
df_average_year = df.groupby('year')['happiness_score'].mean().reset_index()

plt.bar(x = df_Singapore['year'], 
        height = df_Singapore['happiness_score'],
         color = 'b', edgecolor = 'k',align='center')

plt.plot(df_average_year['year'], df_average_year['happiness_score'], marker='o', color='r', label='Average')

plt.yticks(list(range(5,8)));
plt.xticks(df_Singapore['year'], rotation=30, ha='right')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.title('Happiness Score of Singapore')
plt.ylim(5,8)
plt.legend()
```

```python
#Line Chart of Developed Countries

developed_countries = ['United States', 'United Kingdom', 'Singapore', 'Japan', 'Switzerland']
developed_data = df[df['Country Name'].isin(developed_countries)]

plt.figure(figsize=(10, 6))
for country in developed_countries:
    country_data = developed_data[developed_data['Country Name'] == country]
    plt.plot(country_data['year'], country_data['happiness_score'], label=country)
    
plt.plot(df_average_year['year'], df_average_year['happiness_score'], marker='o', color='r', label='Average')

plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.title('Happiness Score Changes in Developed Countries (5 selected)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Line Chart of Developing Countries

developing_countries = ['China', 'Russia','India', 'Brazil', 'South Africa']
developing_data = df[df['Country Name'].isin(developing_countries)]

plt.figure(figsize=(10, 6))
for country in developing_countries:
    country_data = developing_data[developing_data['Country Name'] == country]
    plt.plot(country_data['year'], country_data['happiness_score'], label=country)
    
plt.plot(df_average_year['year'], df_average_year['happiness_score'], marker='o', color='r', label='Average')

plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.title('Happiness Score Changes in Developing Countries (5 selected)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
